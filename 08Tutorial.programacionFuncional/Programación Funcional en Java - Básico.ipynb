````java
Pintable strPintable = () -> "ROJO";
Pintable colorPintable = () -> java.awt.Color.RED;
Pintable hexPintable = () -> "#FF0000";

Collection<Pintable> pintables = List.of(strPintable, colorPintable, hexPintable);

pintables.stream()
    .map(Pintable::pintar)
    .forEach(System.out::println);
````
